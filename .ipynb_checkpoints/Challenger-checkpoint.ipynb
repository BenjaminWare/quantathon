```python
import pandas as pd
import numpy as np
from statsforecast import StatsForecast
from datetime import datetime
import matplotlib.pyplot as plt

import math
from datetime import date, timedelta
from statsforecast.models import MSTL, AutoARIMA,HistoricAverage
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
%env NIXTLA_ID_AS_COL=True
```

```python
delta = pd.read_csv('data/DAL_data.csv')
oil = pd.read_csv('data/CLF_data.csv')

oil_features = pd.read_feather('oil_features')
```

```python
def close_data_to_nixtla_data(df):
    df = df.copy()
    df.reset_index()
    df['ds']=df.index
    df = df.rename(columns={"Adj Close":"y"})
    #Unique_id is to identify which time series we are in
    df['unique_id'] = 1
    return df
```

```python
df = close_data_to_nixtla_data(delta)
```

```python
df2 = close_data_to_nixtla_data(oil)
```

```python
models = [
        # MSTL(
        #     # Seaon length 7 comes from Auotcorrleation function identifying weekly correlations
        #     season_length=[7],
        #     trend_forecaster=AutoARIMA()
        # ),
        HistoricAverage(),
        AutoARIMA()
    ]
    
sf = StatsForecast(models = models, freq='D')
```

```python
df['ds'] = pd.to_datetime(delta['Date']).dt.date
df2['ds'] = pd.to_datetime(oil['Date']).dt.date
```

```python
df = df[['y', 'ds', 'unique_id']]
df2 = df2[['y', 'ds', 'unique_id']]
df2.index = df2.ds
```

```python
oil_features.isna()
```

```python
features_df = df2.join(oil_features).drop(columns=['Date']).iloc[:-48].dropna()
```

```python
features_df.isna().sum()
```

```python
h = 1
cv_df = sf.cross_validation( df=features_df,    
                            h = h,
                            step_size = 1,
                            n_windows = 200)
```

```python
h = 1
cv_df_no_features = sf.cross_validation( df=df2,    
                            h = h,
                            step_size = 1,
                            n_windows = 200)
```

```python
df2.dtypes
```

```python
sf.fit(df=features_df)
```

```python
sf.forecast(h=1, level=[95])
```

```python
cv_df_no_features = cv_df_no_features.drop(['ds','cutoff'],axis=1)
```

```python
cv_df[['y','AutoARIMA','HistoricAverage']].plot()
```

```python
model_mae = np.abs(cv_df['y'] - cv_df['AutoARIMA']).sum()
hist_avg_mae = np.abs(cv_df['y'] - cv_df['HistoricAverage']).sum()
print("Model: ",model_mae)
print("Historic Average: ",hist_avg_mae)
print("Ratio: ", hist_avg_mae/model_mae)
```

```python
cv_df_no_features[['y','AutoARIMA','HistoricAverage']].plot()
```

```python
model_mae = np.abs(cv_df_no_features['y'] - cv_df_no_features['AutoARIMA']).sum()
hist_avg_mae = np.abs(cv_df_no_features['y'] - cv_df_no_features['HistoricAverage']).sum()
print("Model: ",model_mae)
print("Historic Average: ",hist_avg_mae)
print("Ratio: ", hist_avg_mae/model_mae)
```
