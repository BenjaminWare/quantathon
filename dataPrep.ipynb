```python
import pandas as pd
import yfinance as yf
```

```python
spy = yf.download("spy", start="2014-03-24").rename(columns={'Adj Close': 'spy'})
dji = yf.download("^dji", start="2014-03-24").rename(columns={'Adj Close': 'dji'})
ndq = yf.download('^IXIC', start="2014-03-24").rename(columns={'Adj Close': 'ndq'})
eur = yf.download('eur=x', start="2014-03-24").rename(columns={'Adj Close': 'eur'})
```

```python
henryhub = pd.read_csv('exo_data/DHHNGSP.csv')
brentprice = pd.read_csv('exo_data/brent.csv')
wtiprice = pd.read_csv('exo_data/wti.csv')
fedfund = pd.read_csv('exo_data/fedfund.csv')
dollarindex = pd.read_csv('exo_data/realbroaddollarindex.csv')
```

```python
henryhub.index = pd.to_datetime(henryhub['DATE'])
henryhub = henryhub.drop(columns=['DATE'])
henryhub = henryhub[henryhub['DHHNGSP'] != '.'].astype('float64')
brentprice.index = pd.to_datetime(brentprice['DATE'])
brentprice = brentprice.drop(columns=['DATE'])
brentprice = brentprice[brentprice['DCOILBRENTEU'] != '.'].astype('float64')
wtiprice.index = pd.to_datetime(wtiprice['DATE'])
wtiprice = wtiprice.drop(columns=['DATE'])
wtiprice = wtiprice[wtiprice['DCOILWTICO'] != '.'].astype('float64')
fedfund.index = pd.to_datetime(fedfund['DATE'])
fedfund = fedfund.drop(columns=['DATE'])
dollarindex.index = pd.to_datetime(dollarindex['DATE'])
dollarindex = dollarindex.drop(columns=['DATE'])
dollarindex = dollarindex.resample('D').ffill()
```

```python
usoilprod = pd.read_excel('exo_data/usOIL.xls', sheet_name='Data 1')
```

```python
usoilprod.columns = usoilprod.iloc[1]
usoilprod = usoilprod.iloc[2:]
usoilprod.columns
usoilprod = usoilprod[usoilprod['Date'] > pd.to_datetime('2014-03-01')]
usoilprod.index = pd.to_datetime(pd.to_datetime(usoilprod['Date']).dt.date)
```

```python
usoilprod = usoilprod.resample('D').ffill()[['U.S. Field Production of Crude Oil (Thousand Barrels per Day)', 'Date']]
```
